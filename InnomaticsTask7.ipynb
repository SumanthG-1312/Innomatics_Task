```python
# Problem Statement 1 : Create a class for Smart Light
class SmartLight:
    # Constructor runs automatically when object is created
    def __init__(self, name):
        self.name = name        # Store the light name
        self.status = "OFF"     # Default status is OFF
    # Method to turn the light ON
    def turn_on(self):
        self.status = "ON"      # Change status to ON
    # Method to turn the light OFF
    def turn_off(self):
        self.status = "OFF"     # Change status to OFF
    # Method to display current status
    def display_status(self):
        print(f"{self.name} is {self.status}")  # Print name and status
# Create object
light = SmartLight("Bedroom Light")
# Turn the light ON
light.turn_on()
# Display status
light.display_status()
```

```python
# Problem 2: Employee ID Card System

class EmployeeIDSystem:
    def __init__(self,emp_name , emp_id , department):
        self.emp_name = emp_name
        self.emp_id = emp_id
        self.department = department 
    def emp_details(self):
        print("employee ID Card: ")
        print("Employee Name:" ,self.emp_name)
        print("Employee Id:" , self.emp_id)
        print("Department:" , self.department)
employee = EmployeeIDSystem("Sumanth" , "13","AI ML")    
employee.emp_details()
```

```python
# Problem 3: Mobile Contact Record

class MobileContactAppicationss:
    def __init__(self ,name, phone_number):
        self.name = name 
        self.phone_number = phone_number
    def contact_details(self):
        print("Contact Saved")
        print("Name :",self.name)
        print("Phone :",self.phone_number)
contact = MobileContactAppicationss("Sumanth","7392979107")
contact.contact_details()
```

```python
# Problem 4 : Product Price Tag Generator
class PriceTagGenerator:
    def __init__(self , product_name , price):
        self.product_name = product_name
        self.price = price
    def Pricetag(self):
        print("Product :",self.product_name)
        print("Price :",self.price)
p = PriceTagGenerator("Earphones" , 599)
p. Pricetag()
```

```python
# Problem 5: Movie Rating Display System

class MovieRating:
    def __init__(self,movie_name , rating):
        self.movie_name = movie_name
        self.rating = rating 
    def MovieRatting(self):
        print("Movie Name :",self.movie_name)
        print("Rating :",self.rating)
movie1 = MovieRating("Kumbalangi Nights","4.5/5")
movie1.MovieRatting()
```

```python
# Problem 6: Delivery Address Manager

class DeliveryAddressManager:
    def __init__(self , name , address):
        self.name = name 
        self.address = address
    def details (self):
        print("Customer Name :",self.name)
        print("Address :",self.address)
customer = DeliveryAddressManager("Sumanth","Hyderabad")
customer.details()
```
