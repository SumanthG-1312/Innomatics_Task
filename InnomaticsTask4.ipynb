```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]  # these are only allowed recharge plans
    if amount < 50:
        return False   # recharge should not be less than 50
    if amount not in valid_plans:
        return False   # if plan is not in given list, it is invalid
    return True   # if both conditions pass, recharge is valid
while True:
    amount = int(input("Enter recharge amount: "))

    if validate_recharge(amount):
        print("Recharge Successful")
        break   # stopping retry once valid recharge happens
    else:
        print("Invalid Recharge, Try Again")
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for item in products:   # going through each product in dictionary
        if products[item] < 15:
            print(item, ": Reorder Alert")   # if stock is less than 15, need to reorder
        else:
            print(item, ": Stock OK")   # if stock is enough, no action needed
products = {"Laptop": 10, "Mouse": 25, "Keyboard": 12}
check_inventory(products)
```

```python
# Problem 3: Student Result Processing System


def check_result(marks):
    total = 0
    for mark in marks:
        total = total + mark   # adding all marks to calculate average
    average = total / len(marks)   # finding average marks
    if average >= 50:
        return "Pass"   # student passes if average is 50 or more
    else:
        return "Fail"   # otherwise student fails
marks = [60, 45, 70, 50]
print(check_result(marks))
```

```python
# Problem 4: Cab Fare Estimator with Retry Option


def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)   # base fare is 50 and 12 per km
    if peak == "yes":
        fare = fare + (fare * 0.25)   # adding 25% extra during peak hour
    return fare   # returning final calculated fare
while True:
    km = float(input("Enter distance: "))
    peak_hour = input("Peak hour? (yes/no): ")
    print("Total Fare:", calculate_fare(km, peak_hour))
    retry = input("Retry? (yes/no): ")
    if retry == "no":
        break   # stopping loop if user does not want retry
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1   # counting how many days student was present
    percentage = (present / len(attendance)) * 100   # calculating attendance percentage
    if percentage >= 75:
        return "Eligible"   # student is eligible if attendance is 75% or more
    else:
        return "Not Eligible"
attendance = ["P", "P", "A", "P", "P"]
print(check_attendance(attendance))
```

```python
# Problem 6: Password Strength Checker


def check_password(password):
    has_digit = False
    has_special = False#
    if len(password) < 8:
        return "Weak Password"   # password should be at least 8 characters
    for char in password:
        if char.isdigit():
            has_digit = True   # checking if at least one digit exists
        if char in "@#$":
            has_special = True   # checking if at least one special character exists
    if has_digit and has_special:
        return "Strong Password"   # strong only if both conditions satisfied
    else:
        return "Weak Password"
password = input("Enter password: ")
print(check_password(password))
```
