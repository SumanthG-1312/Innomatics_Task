```python
# Problem 1: Unique Words in a Sentence

x = " python is easy and python is powerful" #input string values 
words = x.split() # splitting each  word in the given input 
unique_words = set(words) # we can directly use SET to get unique values because, SET doesn't accept duplicate values (or) repeating words 
print("Count:",len(unique_words)) # len function will help to find the counnt total number of strings in the derived output
print("Unique Words:",unique_words)
```

```python
# Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_salary = 0 # let's take highest as 0 first because whenever we will be moving from the first iteration the first value would become the highest and comparission starts from there we can also take frist value also the highest value 
for i in employees :# traverse through everry value (keys)
    if(employees[i]>highest_salary): # compares the present iteration value with the highest value
        highest_salary = employees[i] # is the values is greater than the highest value then the present value will be replaced by that value.
print("Highest Salary:",highest_salary)
```

```python
# Problem 3: Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
maximum = 45 # initialize it with the first value because if we get the second value greather than this we will repkace it with that vale 
minimum = 45 # initialize it with the first value because if we get the second value less than this we will repkace it with that vale 
for i in numbers : # iteration 
    if (i>maximum): #maximum loigc , if present teration value is greather than then the present maximum value will be replaced by present iteration
        maximum = i
    elif(i<minimum): #maximum loigc , if present teration value is less than then the present minimum value will be replaced by present iteration
        minimum = i
print("Mximum :",maximum)
print("Minimum :",minimum)
```

```python
# Problem 4: Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]
count = 0 # number of counts = 0
for i in prices :
    if(i>1000): # if prices above 1000 found then count will be increased
        count+=1
print("Products above 1000:",count)
```

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]  # list storing attendance status (P = Present, A = Absent)
count = 0          # this will count total number of presents
word = "P"         # we are checking for present days
for i in attendance:     # loop through each attendance record
    if i == word:        # if the day is present
        count += 1       # increase present count
# calculate percentage using formula (present / total days) * 100
total_percentage = count / len(attendance) * 100
print("Attendance Percentage:", total_percentage)
```

```python
# Problem 7: Count Character Frequency

x = "pythonp"      # input string
frequency = {}     # empty dictionary to store character count
for i in x:        # loop through each character in string
    if i in frequency:     # if character already exists in dictionary
        frequency[i] += 1  # increase its count
    else:
        frequency[i] = 1   # if not present, add it with count 1
print(frequency)
```

```python
# Problem 10: Calculate Average Marks

marks = [70, 75, 80, 65]
total = 0
count = 0
for mark in marks:
    total += mark
    count += 1
average = total / count
print("Average Marks:", average)
```
