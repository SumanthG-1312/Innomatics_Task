```python
# Problem Statement 1:Smart Parking Lot Management System

def parking_availabiity(parking_Capacity, vehicle_Logs):
    currently_parked_vehicles = 0
    for i in vehicle_Logs:
        if i == "IN":
            currently_parked_vehicles += 1
        elif i == "OUT":
            currently_parked_vehicles -= 1
    print("Currently parked Vehicles :",currently_parked_vehicles)
    if currently_parked_vehicles < parking_Capacity:
        print("Parking States Available")
    else:
        print("Parking Status : Full")

vehicle_Logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_Capacity = 50
parking_availabiity(parking_Capacity,vehicle_Logs)
```

```python
#  Problem Statement 2:Online Food Delivery Time Estimator

def food_delivery_system(distance,traffic,weather):
    preperation_time = 25 
    if traffic == "High":
        speed = 18
    elif traffic == "Medium":
        speed = 25
    else :
        speed = 30
    travel_time = (distance/speed)*60
    if weather == "Rainy":
        weather = 4
    elif weather == "Storm":
        weather = 8
    else :
        weather = 0
    total_time = preperation_time + travel_time + weather 
    return round(total_time)
print("Estimated Delivery Time:",food_delivery_system(8,"High","Rainy"))
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

total_seats = 200
booked_seats = [1] * 150   # 150 booked seats
def seat_Occupancy_Analyzer(total_seats , booked_seats):
    count = len(booked_seats)

    occupancy = (count / total_seats) * 100

    if occupancy < 50:
        status = "Seats Available"
    elif occupancy < 75:
        status = "Filling Fast"
    elif occupancy < 100:
        status = "Almost Full"
    else:
        status = "Full"

    print("Occupancy:", occupancy, "%")
    print("Show Status:", status)
total_seats = 200
booked_seats = [1]*150
seat_Occupancy_Analyzer(total_seats , booked_seats)
```

```python
#  Problem Statement 4:Cloud Server Load Classification System

def cslcs(cpu_readings):
    total_readings = 0
    for i in cpu_readings:
        total_readings += i        
    average = total_readings / len(cpu_readings)
    if average < 50:
        print("Normal")
    elif 50 <= average < 80:
        print("Warning")
    else:
        print("Critical")
cpu_readings = [45, 60, 70, 85, 90]
cslcs(cpu_readings)
```

```python
#  Problem Statement 5:Smart Classroom Resource Usage Monitor

def smart_classroom_monitor(resource_usage):
    overused = []
    if resource_usage["Projector"] > 6:
        overused.append("Projector")

    if resource_usage["AC"] > 8:
        overused.append("AC")

    if resource_usage["Lights"] > 5:
        overused.append("Lights")

    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
smart_classroom_monitor(resource_usage)
```
