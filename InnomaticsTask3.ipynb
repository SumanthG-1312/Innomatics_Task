```python
#Problem Statement 1: Employee Performance Bonus Eligibility

employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
} # employyes data to check top score 
top_scoring = 0
top_Scorer = []# initially the top score would be 0 , when we traverse through to first index of the data , that value would be highest , in that case it will be easy to comapre the values in the upcoming Iterations 
for i in employees : # loop to traverse from first index to last index 
    if (employees[i] >= top_scoring) : # this is the main part , we check if the present Iteration value is greather than the maximum score .
        top_Scorer.append(i) # if yes the maximum score value will be replaced by the present maximum values with presnt Iteration Value . if no the iteration will continue 
        top_scoring = employees[i]
top = ",".join(top_Scorer) #join is used to convert list of strings into one string 
print(f"Top performerEligible for Bomus : {top} (score:{top_scoring})")
```

```python
# Problem Statement 2: Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"
query = query.lower() # coverts uppercase to lower case 
words = query.split() # splits the words 
frequency = {} # store the frequency of words 
for word in words : # Iteration in words so that every splitted word can be traversed for frequency check
    if word in frequency : # if the word is already in frequency then it will increase the count 
        frequency[word]+=1
    else :
        frequency[word] = 1 # else the count would be 1
result = {} # store the words repeated more than once 
for word in frequency :# traversing in freaquency to find repeated values 
    if frequency[word] > 1: # if te word repeated more than once then , 
        result[word] = frequency[word] # the word will be added to result else the value will be skipped
print(result)
```

```python
# Problem Statement 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:  # checking if reading is even because only even values are valid
        valid_readings.append((i, sensor_readings[i]))  # storing hour index and reading together as a pair
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Problem Statement 4: Email Domain Usage Analysis

emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]  # extracting domain part after '@'
    if domain in domain_count:
        domain_count[domain] += 1  # increasing count if domain already exists
    else:
        domain_count[domain] = 1   # first time seeing this domain
total_users = len(emails)
for domain in domain_count:
    percentage = (domain_count[domain] / total_users) * 100  # calculating usage percentage
    print(f"{domain}: {int(percentage)}%")
```

```python
# Problem Statement 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]
total = 0
for value in sales:
    total = total + value   # adding each sale to find total
average = total / len(sales)   # finding average sales
limit = average + (average * 0.30)   # calculating 30% above average
day = 1   # starting day count from 1
for value in sales:
    if value > limit:   # checking if sale is more than 30% above average
        print("Day", day, ":", value)
    day = day + 1   # moving to next day
```
